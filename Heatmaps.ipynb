```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('2012(heatmap).csv')

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.set_index('hour_of_day'), annot=True, cmap="viridis")
plt.title("Heatmap of Hourly Values by Day (2012)")
plt.xlabel("Day")
plt.ylabel("Hour of Day")
plt.show()
```
